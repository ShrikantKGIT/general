```python
# Use this class to implement your data structure
class Munchlax:
    #CUSTOM_TEST_ARGS = []  # List for custom arguments
    # Use this to initialize any data structures you need
    def __init__(self):
        self.AVLset_weight = AVLSet(key = lambda x: x.weight) #added
        self.AVLset_ID = AVLSet(key = lambda x: x.ID) #added

    # Registers a contestant with the given ID, weight, and rating
    # If contestant is already registered, instead updates weight and rating
    # ID :: int
    # weight :: int
    # rating :: int
    # returns nothing
    def register(self, ID, weight, rating):
        new_contestant = Contestant(ID, weight, rating)
        print('--------###################')
        print('in register...', new_contestant.ID)
        if self.AVLset_ID.find(new_contestant.ID) == None:
            print("not registered")
            self.AVLset_weight.insert(new_contestant)
            print("inserted into weight")
            self.AVLset_ID.insert(new_contestant)
            print("inserted into ID")
        else:
            print("registered*******************************")
            #existing = self.AVLset_weight.find(new_contestant.weight)
            existing = self.AVLset_ID.find(new_contestant.ID)
            existing.rating = rating
            existing.weight = weight

    # Withdraws registration of contestant with the given ID (if any)
    # ID :: int
    # returns nothing
    def withdraw(self, ID):
        new_contestant = Contestant(ID)
        if self.AVLset_ID.find(new_contestant.ID) != None:
            self.AVLset_ID.delete(new_contestant.ID)
            self.AVLset_weight.delete(new_contestant.weight)

    # Among registered contestants with weights >= minweight and <= maxweight,
    # finds and outputs the ID of the contestant with the highest rating
    # minweight :: int
    # maxweight :: int
    # return :: int

    #def best_in_class(self, minweight, maxweight):
    #  return 0
    #    min_recursion = []
    #    #zz = self.AVLset_weight
    #    #print('here....1: ', zz)
    #    #print('here....2: ', minweight)
    #    if self.AVLset_weight > minweight:
    #        return max()


    '''
    def best_in_class(self, minweight, maxweight):
            #if self.AVLset_weight.size():  # Handle empty AVLSet
            #   return 0
            max_rating = float('-inf')  # Initialize with a negative infinity
            max_rating_id = 0

            # Use in_order traversal to iterate through the AVLSet by weight
            for contestant in in_order(self.AVLset_weight.T):  # Access the tree structure

                print('minweight ', minweight)
                print('contestant.item.weight ', contestant.item.weight)
                print('maxweight ', maxweight)

                if minweight <= contestant.item.weight <= maxweight:
                    if contestant.item.rating > max_rating:
                        max_rating = contestant.item.rating
                        max_rating_id = contestant.item.ID
            print('max_rating_id ', max_rating_id)
            return max_rating_id
    '''

    def best_in_class(self, minweight, maxweight):
        # Wrapper function for test cases
        return self.best_in_class_recursive(self.AVLset_weight.T, minweight, maxweight)


    def best_in_class_recursive(self, node, minweight, maxweight):
        if node is None:
            return 0

        max_rating = float('-inf')  # Initialize max_rating
        max_rating_id = 0  # Initialize max_rating_id

        '''
        if minweight <= node.item.weight <= maxweight:
            if node.item.rating > max_rating:
                max_rating = node.item.rating
                max_rating_id = node.item.ID

        left_result = self.best_in_class_recursive(node.left, minweight, maxweight)
        if left_result > max_rating_id:
            max_rating_id = left_result

        print('left_result....', left_result)

        right_result = self.best_in_class_recursive(node.right, minweight, maxweight)
        if right_result > max_rating_id:
            max_rating_id = right_result

        print('right_result....', right_result)
        '''
        # Recursive in-order traversal
        if node.left:
            left_max_rating_id = self.best_in_class_recursive(node.left, minweight, maxweight)

        if minweight <= node.item.weight <= maxweight:
            if node.item.rating > max_rating:
                max_rating = node.item.rating
                max_rating_id = node.item.ID
            elif node.item.rating == max_rating:
                max_rating_id = left_max_rating_id

        if node.right:
            right_max_rating_id = self.best_in_class_recursive(node.right, minweight, maxweight)
            if right_max_rating_id != 0:
                max_rating_id = right_max_rating_id

        print('in best_in_class_recursive max_rating_id....', max_rating_id)
        return max_rating_id





# You may find this container class helpful; feel free to use and/or modify it!
class Contestant:
    cont_counter = 0

    def __init__(self, ID = 0, weight = 0, rating = 0):
        print('in contestant...', ID, weight, rating)
        self.ID = ID
        self.weight = weight
        self.rating = rating
        self.cont_counter += 1

# Below are implementations of the Set and Sequence interfaces using AVL trees
class AVLSet:
    def __init__(self, key = lambda x: x):
        self.key = key
        self.T = None
    def find(self, x):
        T = find(self.T, x, self.key)
        return None if T is None else T.item
    def insert(self, x):
        self.T = insert(self.T, x, self.key)
    def delete(self, x):
        v, self.T = delete(self.T, x, self.key)
        return v

class AVLSeq:
    def __init__(self):
        self.T = None
    def get_at(self, i):
        return get_at(self.T, i)
    def set_at(self, i, x):
        self.T = set_at(self.T, i, x)
    def insert_at(self, i, x):
        self.T = insert_at(self.T, i, x)
    def delete_at(self, i):
        v, self.T = delete_at(self.T, i)
        return v

# Basic tree functionality
class Tree:
    def __init__(self, item, left, right):
        self.item = item
        self.left = left
        self.right = right

# Augmentations
def size(T):
    if T is None:
        return 0
    if not hasattr(T, "size"):
        T.size = 1 + size(T.left) + size(T.right)
    return T.size
def height(T):
    if T is None:
        return -1
    if not hasattr(T, "height"):
        T.height = 1 + max(height(T.left), height(T.right))
    return T.height
def skew(T):
    if T is None:
        return 0
    return height(T.right) - height(T.left)

# AVL functionality
def AVL(T):
    # If T is an AVL tree, outputs T
    # Otherwise outputs an AVL tree T' such that:
    # T' contains the same elements as T in the same order, and
    # height(T) - 1 <= height(T') <= height(T)
    # All subtrees of T except possibly T itself must be AVL trees
    # T must have skew -2, -1, 0, 1, or 2
    if skew(T) < -1:
        if skew(T.left) > 0:
            T = Tree(T.item, left_rotate(T.left), T.right)
        T = right_rotate(T)
    elif skew(T) > 1:
        if skew(T.right) < 0:
            T = Tree(T.item, T.left, right_rotate(T.right))
        T = left_rotate(T)
    return T
def left_rotate(T):
    # Outputs the tree T' that results from a left-rotation at the root of T
    return Tree(T.right.item, Tree(T.item, T.left, T.right.left), T.right.right)
def right_rotate(T):
    # Outputs the tree T' that results from a right-rotation at the root of T
    return Tree(T.left.item, T.left.left, Tree(T.item, T.left.right, T.right))

# Traversals
def in_order(T):
    if T:
        yield from in_order(T.left)
        yield T
        yield from in_order(T.right)
def pre_order(T):
    if T:
        yield T
        yield from pre_order(T.left)
        yield from pre_order(T.right)
def post_order(T):
    if T:
        yield from post_order(T.left)
        yield from post_order(T.right)
        yield T

# Sequence operations
def get_at(T, i):
    # Outputs the subtree of T whose root is the element at index i in T
    if i < size(T.left):
        return get_at(T.left, i)
    if i > size(T.left):
        return get_at(T.right, i - 1 - size(T.left))
    return T
def set_at(T, i, x):
    # Outputs the tree that results from replacing the element at index i in T with x
    # i must be in the range 0 <= i < size(T)
    if i < size(T.left):
        return Tree(T.item, set_at(T.left, i, x), T.right)
    if i > size(T.right):
        return Tree(T.item, T.left, set_at(T.right, i - 1 - size(T.left), x))
    return Tree(x, T.left, T.right)
def insert_at(T, i, x):
    # Outputs the AVL tree that results from inserting x at index i in T
    # i must be in the range 0 <= i <= size(T)
    if T is None:
        return Tree(x, None, None)
    if i > size(T.left):
        return AVL(Tree(T.item, T.left, insert_at(T.right, i - 1 - size(T.left), x)))
    return AVL(Tree(T.item, insert_at(T.left, i, x), T.right))
def delete_at(T, i):
    # Outputs a pair (x, T'), where T' is the AVL tree that results from deleting x at index i in T
    # i must be in the range 0 <= i < size(T)
    if i < size(T.left):
        rec = delete_at(T.left, i)
        return (rec[0], AVL(Tree(T.item, rec[1], T.right)))
    if i > size(T.left):
        rec = delete_at(T.right, i - 1 - size(T.left))
        return (rec[0], AVL(Tree(T.item, T.left, rec[1])))
    if not T.left:
        return (T.item, T.right)
    if not T.right:
        return (T.item, T.left)
    rec = delete_at(T.right, 0)
    return (T.item, AVL(Tree(rec[0], T.left, rec[1])))

# BST operations
def find(T, x, key = lambda i: i):
    # Outputs the subtree of T whose root has key x
    # T must have BSP w.r.t. key
    if not T:
        return None
    if x < key(T.item):
        return find(T.left, x, key)
    if x > key(T.item):
        return find(T.right, x, key)
    return T
def insert(T, x, key = lambda i: i):
    # Outputs the AVL tree that results from adding x to T
    # T must have BSP w.r.t. key
    if not T:
        return Tree(x, None, None)
    if key(x) < key(T.item):
        return AVL(Tree(T.item, insert(T.left, x, key), T.right))
    if key(x) > key(T.item):
        return AVL(Tree(T.item, T.left, insert(T.right, x, key)))
    return Tree(x, T.left, T.right)
def delete(T, x, key = lambda i: i):
    # Outputs the pair (T', y) where y is the item (if any) in T with key x, and T' is the AVL tree that results from deleting y from T
    # T must have BSP w.r.t. key
    if not T:
        return (None, T)
    if x < key(T.item):
        rec = delete(T.left, x, key)
        return (rec[0], AVL(Tree(T.item, rec[1], T.right)))
    if x > key(T.item):
        rec = delete(T.right, x, key)
        return (rec[0], AVL(Tree(T.item, T.left, rec[1])))
    return delete_at(T, size(T.left))
```
